```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
pd.options.display.max_columns=999
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train= pd.read_csv('/kaggle/input/DontGetKicked/training.csv')
test = pd.read_csv('/kaggle/input/DontGetKicked/test.csv')
```

```python
display(train,test)
```

```python
# category 형이 많다.
```

```python
all_data = pd.concat([train,test])
all_data.head()
```

```python
all_data.info()
```

```python
all_data.columns
```

```python
all_data2 = all_data.drop(['RefId', 'IsBadBuy','PurchDate'],axis=1)
all_data2.head()
```

```python
all_data2['Auction'].unique()
```

```python
all_data2['Auction'] = all_data2['Auction'].replace({'ADESA':0,'OTHER':1,'MANHEIM':2})
all_data2.head(1)
```

```python
all_data2['Make'].unique()
```

```python
# 자동으로 바꾸는 library import하기.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#등록 후 변환까지.
all_data2['Make'] = le.fit_transform(all_data2['Make'])
all_data2
```

```python
le.classes_
```

```python
# 1 apply를 사용하여 labelEncoder
# 2. 반복문을 활용한다.
all_data2.columns
```

```python
# 문자형만 관심이 있음.
c = all_data2.columns[all_data2.dtypes==object]
```

```python
for i in c:
    all_data2[i] = le.fit_transform(list(all_data2[i]))
    
# NaN이 category 형태로 변화해서 오류가 생기지 않는다.
```

```python
all_data2.head()
```

```python
all_data2=all_data2.fillna(0)
```

```python
train2=all_data2[:len(train)]
test2 =all_data2[len(train):]
```
